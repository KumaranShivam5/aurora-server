This notebook is for extracting RA and DEC information from Catalog files and storing them as CSV filesm

```python
from astropy.io import fits
import pandas as pd 
from matplotlib import pyplot as plt 
from IPython.display import display
```

```python
hdu = fits.open('cat/AGN_veroncat.fits')[1]
data = hdu.data
df = pd.DataFrame({
    "name" : data['NAME'] , 
    "ra" : data['RA'] , 
    "dec" : data['DEC'],
    "radio_detect" : data['RADIO_DETECT'] , 
    "AGN_type" : data['BROAD_TYPE']
})
display(df)
df.to_csv('AGN_veroncat.csv')
```
