## Cross match AGN veroncat catalogue

```python
df_agn = pd.read_csv('cat/AGN_veroncat.csv', index_col=0)
df_agn.index.name = 'index'
display(df_agn)
agn_coords = SkyCoord(
        ra = df_agn['ra']*u.deg , 
        dec = df_agn['dec']*u.deg
            )
```

```python
from matplotlib import cm
plt.figure(figsize=(20,12))
plt.subplot(111, projection='aitoff' ,)
#plt.subplot(111)
plt.grid(True)



eq = SkyCoord(df['ra'] , df['dec'] , unit = u.deg)
gal = eq.galactic
plt.scatter(gal.l.wrap_at('180d').radian, gal.b.radian ,  s=0.5 , marker='.' , color='k' , alpha=0.2)

eq = SkyCoord(df_agn['ra'] , df_agn['dec'] , unit = u.deg)
gal = eq.galactic
plt.scatter(gal.l.wrap_at('180d').radian, gal.b.radian , s=0.5 , marker='.' , color='r' , alpha=0.2)
#plt.hexbin(gal.l.wrap_at('180d').radian, gal.b.radian ,mincnt=1)
#plt.colorbar()
#plt.scatter(df['ra'] , df['dec'] , alpha=0.2, s=1)
plt.title('CSC 2.0 variable Point sources')

plt.show()
```

```python
idx_ , sep2d_ , _ =  csc_coords.match_to_catalog_sky(
    agn_coords
)
```

```python
plt.hist(sep2d_.arcsec, histtype='step',
         bins=np.logspace(-2, 2., 64))
plt.xlabel('separation [arcsec]')
plt.xscale('log')
plt.yscale('log')
plt.tight_layout()
```

```python
cm = pd.DataFrame({
    'idx' : idx_ , 
    'sep' : sep2d_.arcsec
})
```

```python
for i in range(len(cm)):
    idx = cm['idx'][i]
    sep = cm['sep'][i]
    if(sep<4):
        df.loc[i ,'class'] = 'AGN'
        df.loc[i , 'cat_name'] = df_agn.loc[idx]['name']
        df.loc[i , 'cat_ra'] = df_agn.loc[idx]['ra']
        df.loc[i , 'cat_dec'] = df_agn.loc[idx]['dec']
        df.loc[i , 'offset'] = sep
        df.loc[i , 'catalog'] = 'VERONCAT'
```

```python
df_agn_csc = df[df['class']=='AGN']
df_agn_csc = df_agn_csc.sort_values(by='offset')
df_agn_csc
```

```python
c_names = df_agn_csc['cat_name'].unique()
df_temp = pd.DataFrame()
for c in c_names[:]:
    #df[df['cat_name']==c_names[0]].he
    df_temp = df_temp.append(df_agn_csc[df_agn_csc['cat_name']==c].head(1))
#display(df_temp)
```

```python
print(len(df_temp['cat_name'].unique()))
print(len(df_temp['name'].unique()))
```

```python
df_temp.to_csv('data_v2/csc_agn.csv')
```
