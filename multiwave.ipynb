# NED Query

```python
from astroquery.ipac.ned import Ned
```

```python
#df_selected = df[df['class']=='ULX']
not_done = []
from tqdm import tqdm 
for n in tqdm(df['name'][:]):
    #print(n)
    try:
        result_table = Ned.get_table(n)
        res_df = result_table.to_pandas()
    #display(res_df)
        #res_df.to_csv('multi_wave_data/'+n+'.csv')
        res_df.to_csv('csc_all_mw/'+n+'.csv')
    except Exception as e:
        print(e)
        not_done.append(n)
        print(n)
```

# Vizier query service

```python
from astroquery.vizier import Vizier
from astropy.coordinates import Angle
v = Vizier(catalog='II/246' ,columns= ['RAJ2000',	'DEJ2000',	'2MASS',	'Jmag',	'e_Jmag',	'Hmag',	'e_Hmag',	'Kmag',	'e_Kmag'])
```

```python
df_tmass_ulx = pd.DataFrame()
k = 0
for i , r,d in zip(df_agn['csc_index'] ,  df_agn['ra'] , df_agn['dec']):
    c = coord.SkyCoord(ra=r, dec=-d,unit=(u.deg, u.deg),frame='icrs')
    result = v.query_region(c, radius=Angle(30, "arcsec"),)
    try:
        df_tmass_ulx =  df_tmass_ulx.append(result.values()[0].to_pandas())
    except:
        print(f'{k} , {i} ,match not found')
    k+=1
```
